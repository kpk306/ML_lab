```python
data = [
    ('Sunny', 'Warm', 'Yes'),
    ('Sunny', 'Cold', 'Yes'),
    ('Rainy', 'Cold', 'No')
]

rules = []

for example in data:
    if example[-1] == 'Yes':
        rule = f"IF Weather={example[0]} AND Temp={example[1]} THEN Play=Yes"
        rules.append(rule)

print("Generated Rules:")
for r in rules:
    print(r)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

data = load_iris()
X, y = data.data, data.target

model = DecisionTreeClassifier()
model.fit(X, y)

print("Decision Tree trained successfully")
print("Rules are extracted from root to leaf paths")
```

```python
import numpy as np

Q = np.zeros((3, 2))  # states x actions
alpha = 0.1
gamma = 0.9

state = 0
action = 1
reward = 1
next_state = 1

Q[state, action] += alpha * (reward + gamma * max(Q[next_state]) - Q[state, action])

print("Updated Q-Table:")
print(Q)
```

```python
import random

states = [0,1,2,3]
rewards = {3:10}
policy_reward = 0

for _ in range(10):
    state = random.choice(states)
    policy_reward += rewards.get(state, -1)

print("Total Reward:", policy_reward)
```
