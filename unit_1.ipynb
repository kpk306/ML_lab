```python
# Demonstration of Inductive Bias using Find-S Algorithm

def find_s(data):
    hypothesis = ['0'] * (len(data[0]) - 1)

    for example in data:
        if example[-1] == 'Yes':   # Only positive examples are considered
            for i in range(len(hypothesis)):
                if hypothesis[i] == '0':
                    hypothesis[i] = example[i]
                elif hypothesis[i] != example[i]:
                    hypothesis[i] = '?'
    return hypothesis

training_data = [
    ['Sunny', 'Warm', 'Yes'],
    ['Sunny', 'Cold', 'Yes'],
    ['Rainy', 'Warm', 'No']
]

final_hypothesis = find_s(training_data)
print("Learned Hypothesis:", final_hypothesis)
print("Inductive Bias: Algorithm assumes target concept exists in hypothesis space")
```

```python
import math

def entropy(p, n):
    total = p + n
    if p == 0 or n == 0:
        return 0
    return - (p/total)*math.log2(p/total) - (n/total)*math.log2(n/total)

print("Entropy:", entropy(9, 5))
```
