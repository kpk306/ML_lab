```python
import math
from collections import Counter

def euclidean(a, b):
    return math.sqrt(sum((x-y)**2 for x, y in zip(a, b)))

X_train = [[1,2],[2,3],[3,3],[6,5],[7,7]]
y_train = ['A','A','A','B','B']
test = [3,4]
K = 3

distances = []
for i in range(len(X_train)):
    d = euclidean(X_train[i], test)
    distances.append((d, y_train[i]))

distances.sort()
neighbors = [label for _, label in distances[:K]]
prediction = Counter(neighbors).most_common(1)[0][0]

print("Predicted Class:", prediction)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import numpy as np

def lwr(x0, X, y, tau=1.0):
    X = np.c_[np.ones(len(X)), X]
    x0 = np.array([1, x0])
    W = np.diag(np.exp(-np.sum((X[:,1:] - x0[1:])**2, axis=1)/(2*tau*tau)))
    theta = np.linalg.pinv(X.T @ W @ X) @ X.T @ W @ y
    return x0 @ theta

X = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

print("Prediction:", lwr(3.5, X, y))
```

```python
import numpy as np

def rbf(x, c, sigma):
    return np.exp(-np.linalg.norm(x-c)**2 / (2*sigma**2))

X = np.array([[1],[2],[3],[6],[7]])
y = [0,0,0,1,1]
centers = [2,6]
sigma = 1.0

features = [[rbf(x, c, sigma) for c in centers] for x in X]
print("RBF Features:", features)
```

```python
import math

cases = {
    (25, 50000): "Loan Approved",
    (40, 30000): "Loan Rejected"
}

def similarity(a, b):
    return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)

new_case = (30, 45000)

best_case = min(cases, key=lambda x: similarity(x, new_case))
print("Predicted Decision:", cases[best_case])
```
