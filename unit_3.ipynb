```python
# Bayes Theorem for classification

# Prior probabilities
P_Positive = 0.6
P_Negative = 0.4

# Likelihoods
P_Test_Positive = 0.7
P_Test_Negative = 0.2

# Posterior probabilities
Posterior_Pos = P_Test_Positive * P_Positive
Posterior_Neg = P_Test_Negative * P_Negative

print("Posterior Positive:", Posterior_Pos)
print("Posterior Negative:", Posterior_Neg)

if Posterior_Pos > Posterior_Neg:
    print("Classified as Positive")
else:
    print("Classified as Negative")
```

```python
import numpy as np

data = np.array([4, 6, 8, 10, 12])

mean = np.mean(data)
variance = np.var(data)

print("Mean (MLE):", mean)
print("Variance (MLE):", variance)
```

```python
# Simple Naive Bayes classifier

# Priors
P_Yes = 0.6
P_No = 0.4

# Likelihoods
P_Sunny_Yes = 0.5
P_Sunny_No = 0.2

# Posterior
Yes_prob = P_Sunny_Yes * P_Yes
No_prob = P_Sunny_No * P_No

print("Yes Probability:", Yes_prob)
print("No Probability:", No_prob)

if Yes_prob > No_prob:
    print("Prediction: Yes")
else:
    print("Prediction: No")
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = GaussianNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
