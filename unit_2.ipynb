```python
import numpy as np

# Step function
def step(x):
    return 1 if x >= 0 else 0

# Training data (AND gate)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

w = np.zeros(2)
b = 0
lr = 0.1

for _ in range(10):
    for i in range(len(X)):
        y_pred = step(np.dot(X[i], w) + b)
        error = y[i] - y_pred
        w += lr * error * X[i]
        b += lr * error

print("Weights:", w)
print("Bias:", b)
```

```python
import sys
!{sys.executable} -m pip install --upgrade pip --quiet
!{sys.executable} -m pip install scikit-learn --quiet
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import random

def fitness(x):
    return x*x

population = [random.randint(-10,10) for _ in range(6)]

for _ in range(10):
    population.sort(key=fitness, reverse=True)
    population = population[:4]
    population.append(random.randint(-10,10))
    population.append(random.randint(-10,10))

print("Best solution:", population[0])
```

```python
import random

features = 5
population = [[random.randint(0,1) for _ in range(features)] for _ in range(4)]

def fitness(chromosome):
    return sum(chromosome)

best = max(population, key=fitness)
print("Best Feature Set:", best)
```
